```python
import sys
!{sys.executable} -m pip install pillow
#chargement de la bibliothque PIL permettant la manipulation d'images
from PIL import Image 
#chargement du module display pour l'affichage d'images
from IPython.display import display

# ouverture de l'image qui sera reconnue sous les noms de variable im1. Ce nom est arbitraire.
im1=Image.open("5.3. Lenna.png")
L,H = im1.size

#cration de l'image retouche qui contiendra le texte
im2=im1.copy()

#cration de la liste des modifications des pixels rouges, verts et bleus
r=[0b000,0b000,0b000,0b000,0b000,0b000,0b000,0b000,0b000,0b000]
v=[0b00,0b00,0b00,0b00,0b00,0b00,0b00,0b00,0b00,0b00]
b=[0b000,0b000,0b000,0b000,0b000,0b000,0b000,0b000,0b000,0b000]

#modificaltion des valeurs des pixels
for x in range(9):
    p = im1.getpixel((x,0))
    r1 = p[0]&0b11111000    #mise  zro des 3 derniers pixels
    r2 = r1|r[x]            #Insertion des valeurs binaires sur les 3 derniers pixels
    v1 = p[1]&0b11111100    #mise  zro des 2 derniers pixels
    v2 = v1|v[x]            #Insertion des valeurs binaires sur les 2 derniers pixels
    b1 = p[2]&0b11111000    #mise  zro des 3 derniers pixels
    b2 = b1|b[x]            #Insertion des valeurs binaires sur les 3 derniers pixels
    im2.putpixel((x,0),(r2,v2,b2))
    
    
display(im1)
display(im2)
im2.save("stgano.png")
```
